```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
df = pd.read_csv("/content/heart.csv", sep=",")
print(df.head())
```

```python
df.isnull().sum()
```

```python
# Features (Independent variables)
feature = df[['age', 'chol', 'trestbps']]

# Target (Dependent variable)
Target = df['target']
```

```python
DT = DecisionTreeClassifier(criterion='entropy')

# Train the model
DT.fit(feature, Target)

# Accuracy score
print("Accuracy:", DT.score(feature, Target))
```

```python
# User input
age = int(input("Enter Age: "))
chol = int(input("Enter Cholesterol Level: "))
bp = int(input("Enter Resting Blood Pressure: "))

# Prediction
pred = DT.predict([[age, chol, bp]])

# Output
if pred[0] == 1:
    print("Heart Disease Present")
else:
    print("No Heart Disease")
```

```python
plot_tree(DT) #plot_tree(trained model name)
```
